This function converts the excel file downloaded to a csv file to make it easier to manipulate with pandas.

Note: I have since realized that I could read an excel file directly into pandas using pd.read_excel(). :facepalm:

````python
def csv_from_excel(input_xlsx_filename, specified_xlsx_sheet_name, output_destination_csv_filename):
    """
    This function converts the excel file downloaded to a csv file.
    Inputs include the location of the excel file, the sheet name of the excel file to be converted,
    and the location of the output csv document
    """
    wb = xlrd.open_workbook(input_xlsx_filename)
    sh = wb.sheet_by_name(specified_xlsx_sheet_name)
    output_csv_file = open(output_destination_csv_filename, 'w')
    wr = csv.writer(output_csv_file, quoting=csv.QUOTE_ALL)

    for rownum in range(sh.nrows):
        wr.writerow(sh.row_values(rownum))

    output_csv_file.close()
````
